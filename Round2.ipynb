```python
!pip install transformers --quiet

import pandas as pd
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# === 1. Load your Google Sheet as CSV ===
sheet_url = 'https://docs.google.com/spreadsheets/d/1uOg3_pO4iNXjGfRVHS0qkEC_9TwSdB7tF24WwIYJpis/export?format=csv&gid=1600079014'
df = pd.read_csv(sheet_url)

# === 2. Load the model and tokenizer ===
model_name = "deepseek-ai/deepseek-coder-1.3b-instruct"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

# === 3. Code generation function with token count ===
def generate_code(instruction):
    prompt = f"<|user|>\n{instruction.strip()}\n<|assistant|>\n"
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    token_count = len(inputs.input_ids[0])

    with torch.no_grad():
        output = model.generate(
            **inputs,
            max_new_tokens=200,
            temperature=0.2,
            pad_token_id=tokenizer.eos_token_id
        )

    decoded = tokenizer.decode(output[0], skip_special_tokens=True)
    if "<|assistant|>" in decoded:
        code = decoded.split("<|assistant|>")[-1].strip()
    else:
        code = decoded.strip()

    return code, token_count

# === 4. Process form responses ===
for _, row in df.iterrows():
    team = row.iloc[2]
    instruction = str(row.iloc[3]).strip()

    try:
        code, token_count = generate_code(instruction)
    except Exception as e:
        code = f"Error: {e}"
        token_count = "N/A"

    # === 5. Output nicely ===
    print("=" * 50)
    print(f"Team: {team}")
    print(f"Instruction: {instruction}")
    print(f"Token Count: {token_count}")
    print("-" * 50)
    print("Generated Code:\n")
    print(code)
    print("=" * 50 + "\n")
```

```python
# ✅ Step 1: Install Transformers
!pip install transformers --quiet

# ✅ Step 2: Upload the CSV file manually
from google.colab import files
uploaded = files.upload()

# ✅ Step 3: Load Data
import pandas as pd
from transformers import AutoTokenizer

filename = next(iter(uploaded))  # Gets the uploaded file name
df = pd.read_csv(filename)

# ✅ Step 4: Load a Tokenizer
tokenizer = AutoTokenizer.from_pretrained("gpt2")

# ✅ Step 5: Define prompt columns and corresponding token columns
prompt_cols = {
    'Problem A: Shortest Prompt': 'Tokens A',
    'Problem B: Shortest Prompt': 'Tokens B',
    'Problem C: Shortest Prompt': 'Tokens C',
    'Problem D: Shortest Prompt': 'Tokens D',
    'Problem E: Shortest Prompt': 'Tokens E'
}

# ✅ Step 6: Count tokens per column
for prompt_col, token_col in prompt_cols.items():
    df[token_col] = df[prompt_col].apply(lambda text: 0 if pd.isna(text) else len(tokenizer.tokenize(str(text))))

# ✅ Step 7: Compute total tokens
df['Total Tokens'] = df[list(prompt_cols.values())].sum(axis=1)

# ✅ Step 8: View and export result
result_cols = ['Team Name:'] + list(prompt_cols.values()) + ['Total Tokens']
token_counts = df[result_cols]

display(token_counts)

# Optional: Export
token_counts.to_csv("team_token_counts_detailed.csv", index=False)
files.download("team_token_counts_detailed.csv")
```
